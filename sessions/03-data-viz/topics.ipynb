## Why Visualize Data?

Data visualizations help us see patterns that are difficult to notice in raw tables. A good visualization can help answer questions such as:

- Which category appears most often?
- How are values distributed?
- Is there a relationship between two variables?
- How does something change over time?

Visualization is not just about making charts. It is about making patterns understandable.

</div>

## Matplotlib vs. Seaborn

### Matplotlib
`matplotlib` is the core plotting library in Python. It gives you detailed control over almost every part of a figure.

### Seaborn
`seaborn` is built on top of `matplotlib`. It makes it easier to create attractive statistical graphics with less code.

A useful way to think about them is:

- Use `matplotlib` when you want direct control.
- Use `seaborn` when you want fast, clean statistical plots.
- In practice, many people use both together.

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="deep")
np.random.seed(42)
```

## Create Example Datasets

To keep this notebook self-contained, we will create a few small datasets in code. This means you can run everything without needing to download any files.

```python
# Time-series style data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = pd.DataFrame({
    "month": months,
    "sales": [120, 135, 150, 160, 180, 210],
    "ad_spend": [15, 18, 20, 24, 28, 30]
})

# Categorical data
departments = pd.DataFrame({
    "department": ["Psychology", "Biology", "Cognitive Science", "Computer Science", "Linguistics"],
    "students": [80, 65, 95, 110, 40]
})

# Individual-level data
n = 150
students = pd.DataFrame({
    "study_hours": np.random.normal(12, 3, n).clip(3, 22),
    "exam_score": np.random.normal(78, 10, n).clip(45, 100),
    "group": np.random.choice(["Control", "Intervention"], size=n),
    "major": np.random.choice(["CogSci", "Psych", "CS"], size=n, p=[0.4, 0.35, 0.25])
})
students["passed"] = np.where(students["exam_score"] >= 70, "Yes", "No")

sales.head(), departments.head(), students.head()
```

## The Anatomy of a Plot

Most plots include the same basic parts:

- **Figure**: the whole canvas
- **Axes**: the area where data is drawn
- **Title**: tells the reader what the plot is about
- **X-axis label** and **Y-axis label**: explain the variables
- **Legend**: identifies groups or colors when needed

When a plot is confusing, it is often because one or more of these pieces is missing.

## 1. Line Plots with Matplotlib

Line plots are useful when you want to show **change over time** or across an ordered sequence.

In the example below, we plot monthly sales. Because the months have a natural order, a line plot makes sense.

```python
plt.figure(figsize=(8, 4))
plt.plot(sales["month"], sales["sales"], marker="o", linewidth=2)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### What does this plot show?

The line rises from January to June, which suggests that sales are increasing over time. The markers make each data point easier to see.

## 2. Bar Plots with Matplotlib

Bar plots are useful for comparing **categories**. Each bar represents a group, and the height shows the value for that group.

Use bar plots when you want to answer questions like:

- Which group is largest?
- How do categories compare?
- Which department has the most students?

```python
plt.figure(figsize=(8, 4))
plt.bar(departments["department"], departments["students"], color="steelblue")
plt.title("Students by Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")
plt.xticks(rotation=20)
plt.show()
```

### Why rotate the labels?

Long category names can overlap. Rotating labels makes them easier to read. Small formatting choices like this can make a plot much more usable.

## 3. Histograms with Matplotlib

A histogram shows the **distribution** of a numeric variable. Instead of showing one bar per category, it groups values into bins.

Histograms help answer questions like:

- Are most values clustered together?
- Is the distribution symmetric or skewed?
- Are there extreme values?

```python
plt.figure(figsize=(8, 4))
plt.hist(students["exam_score"], bins=12, color="mediumpurple", edgecolor="black")
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

### Interpreting this histogram

The tallest bars show the score ranges where most students are concentrated. If one side stretches farther than the other, the distribution may be skewed.

## 4. Why Use Seaborn?

Seaborn works especially well with pandas DataFrames because you can refer to columns by name. This makes your code easier to read.

For example, compare these two styles:

- `plt.scatter(students["study_hours"], students["exam_score"])`
- `sns.scatterplot(data=students, x="study_hours", y="exam_score")`

The second version is often easier to understand.

## 5. Scatter Plots with Seaborn

Scatter plots are used to examine the relationship between **two numeric variables**.

Here we ask: Do students who study more tend to score higher?

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=students, x="study_hours", y="exam_score", hue="group")
plt.title("Study Hours and Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

### Interpreting a scatter plot

If the points trend upward from left to right, that suggests a positive relationship. If they trend downward, that suggests a negative relationship. If they look scattered without a pattern, there may be little relationship.

## 6. Adding a Trend Line

Sometimes it helps to add a fitted line to summarize a pattern. Seaborn makes that easy with `regplot()`.

This does not prove causation. It only helps summarize the overall association.

```python
plt.figure(figsize=(8, 5))
sns.regplot(data=students, x="study_hours", y="exam_score", scatter_kws={"alpha": 0.7})
plt.title("Study Hours and Exam Scores with Trend Line")
plt.show()
```

## 7. Box Plots with Seaborn

Box plots are useful for comparing the distribution of a numeric variable across groups.

They show:

- the median
- the middle 50% of the data
- possible outliers

This makes them useful when you want to compare groups such as different majors or treatment conditions.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=students, x="major", y="exam_score")
plt.title("Exam Scores by Major")
plt.xlabel("Major")
plt.ylabel("Exam Score")
plt.show()
```

### How do you read a box plot?

Look at the median line first. Then compare the spread of the boxes. Wider spread suggests more variability. Points beyond the whiskers may be outliers.

## 8. Count Plots with Seaborn

A count plot is a quick way to visualize how many observations fall into each category. It is like a bar plot, but seaborn does the counting for you automatically.

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=students, x="major")
plt.title("Number of Students by Major")
plt.xlabel("Major")
plt.ylabel("Count")
plt.show()
```

## 9. Histograms with Seaborn

Seaborn can also create histograms. One useful option is adding a KDE curve, which gives a smoothed view of the distribution.

Use this carefully: the KDE is an estimate, not the raw data.

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=students, x="exam_score", bins=12, kde=True)
plt.title("Exam Score Distribution with KDE")
plt.show()
```

## 10. Violin Plots

A violin plot is similar to a box plot, but it also shows the shape of the distribution. This can be useful when you want more detail than a box plot provides.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=students, x="major", y="exam_score")
plt.title("Exam Score Distribution by Major")
plt.show()
```

## 11. Heatmaps

Heatmaps use color to represent values in a matrix or table. A common use is displaying a correlation matrix.

Correlation values range from -1 to 1:

- values near 1 suggest a strong positive relationship
- values near -1 suggest a strong negative relationship
- values near 0 suggest a weak linear relationship

```python
numeric_cols = students[["study_hours", "exam_score"]].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(numeric_cols, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

## 12. Customizing Your Plots

Good plots are not only correct. They are also readable.

Common customizations include:

- `figsize=` to change the size
- `title()` to explain the message
- `xlabel()` and `ylabel()` to label axes
- `palette=` or `color=` to change colors
- `legend()` to describe groups
- `plt.tight_layout()` to reduce overlap

```python
plt.figure(figsize=(9, 5))
sns.barplot(data=departments, x="department", y="students", palette="viridis")
plt.title("Department Enrollment", fontsize=14, weight="bold")
plt.xlabel("Department")
plt.ylabel("Students")
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

## 13. Creating Multiple Plots in One Figure

Sometimes you want to compare plots side by side. Subplots let you do that.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(data=students, x="study_hours", bins=12, ax=axes[0], color="teal")
axes[0].set_title("Study Hours")

sns.histplot(data=students, x="exam_score", bins=12, ax=axes[1], color="coral")
axes[1].set_title("Exam Scores")

plt.tight_layout()
plt.show()
```

## 14. Common Visualization Mistakes

Try to avoid these:

- Missing titles or labels
- Using too many colors without meaning
- Choosing the wrong chart type
- Overcrowding a plot with too much information
- Making text too small to read

A simple, readable plot is usually better than a flashy one.
