```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## 1. Line Plot
Good for trends over time.


```python
temps = pd.DataFrame({
    "day": [1, 2, 3, 4, 5],
    "temp": [68, 70, 69, 72, 75]
})

plt.plot(temps["day"], temps["temp"], marker="o")
plt.title("Daily Temperature")
plt.xlabel("Day")
plt.ylabel("Temp (F)")
plt.show()
```

## 2. Bar Chart
Good for comparing categories.


```python
majors = pd.DataFrame({
    "major": ["Bio", "CS", "Math"],
    "students": [40, 55, 25]
})

plt.bar(majors["major"], majors["students"])
plt.title("Students by Major")
plt.xlabel("Major")
plt.ylabel("Students")
plt.show()
```

## 3. Histogram
See distributions.


```python
sns.histplot([2.1, 2.5, 3.0, 3.2, 3.8, 3.9, 2.7])
plt.title("GPA Distribution")
plt.show()
```

## 4. Scatter Plot
See relationships between two numeric variables.


```python
study = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6],
    "score": [60, 65, 70, 78, 85, 90]
})

sns.scatterplot(data=study, x="hours", y="score")
plt.title("Study Hours vs Score")
plt.show()
```
