Imagine we are studying student experiences with AI tools in higher education. The short text responses below resemble the kind of open-ended responses a researcher might collect in a survey.

```python
responses = [
    "AI tools help me brainstorm ideas quickly, but I worry that I rely on them too much.",
    "Using ChatGPT saved me time when I was outlining my essay and organizing my thoughts.",
    "I am concerned that students may stop practicing writing if they depend heavily on AI.",
    "AI makes research faster, especially when I need a quick summary of complex readings.",
    "Sometimes the AI sounds confident but gives inaccurate answers, which makes me cautious."
]

responses
```

## 3. The Basic NLP Workflow

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">How NLP is usually done</h2>
A simple NLP workflow often looks like this:
<ol>
<li>Collect text data</li>
<li>Clean the text</li>
<li>Break text into tokens (words or sentences)</li>
<li>Remove irrelevant words or punctuation</li>
<li>Transform the text into something measurable</li>
<li>Analyze or model the results</li>
</ol>
Each step helps convert human language into a form that computers can work with.
</div>

## 4. Tokenization

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">What is tokenization?</h2>
Tokenization means breaking text into smaller units. These units are usually words or sentences. It is one of the first steps in NLP because most later tasks work on tokens rather than raw text.<br><br>
<b>Social science use:</b> tokenization is useful when analyzing interview transcripts, open-ended survey responses, or speeches because it helps convert those texts into units we can count or compare.
</div>

```python
text = responses[0]

print("Sentence tokenization:")
print(sent_tokenize(text))

print("\nWord tokenization:")
print(word_tokenize(text))
```

## 5. Text Cleaning

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">Why clean text?</h2>
Raw text contains punctuation, capitalization differences, and common words that may not help with analysis. Cleaning text helps reduce noise.<br><br>
Common cleaning steps include:
<ul>
<li>lowercasing</li>
<li>removing punctuation</li>
<li>removing stopwords</li>
</ul>
</div>

```python
stop_words = set(stopwords.words('english'))

tokens = word_tokenize(text.lower())
tokens_alpha = [t for t in tokens if t.isalpha()]
tokens_clean = [t for t in tokens_alpha if t not in stop_words]

print("Original tokens:", tokens)
print("\nAlphabetic tokens:", tokens_alpha)
print("\nClean tokens:", tokens_clean)
```

## 6. Stemming and Lemmatization

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">Stemming vs. lemmatization</h2>
Both stemming and lemmatization try to reduce words to a more basic form.
<ul>
<li><b>Stemming</b> chops off word endings, often roughly</li>
<li><b>Lemmatization</b> tries to recover the proper dictionary form</li>
</ul>
This helps when you want to treat related words like <code>learn</code>, <code>learning</code>, and <code>learned</code> as similar.
</div>

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ["learning", "learned", "studies", "studying", "students"]

stems = [stemmer.stem(w) for w in words]
lemmas = [lemmatizer.lemmatize(w) for w in words]

pd.DataFrame({"word": words, "stem": stems, "lemma": lemmas})
```

## 7. Frequency Analysis

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">Why count words?</h2>
Frequency analysis is one of the simplest NLP techniques. It shows which words appear most often. Even this basic method can be useful in social science. For example, researchers might compare which themes appear most often in political speeches, student reflections, or media coverage.
</div>

```python
all_tokens = []
for response in responses:
    toks = word_tokenize(response.lower())
    toks = [t for t in toks if t.isalpha() and t not in stop_words]
    all_tokens.extend(toks)

freq = FreqDist(all_tokens)
freq.most_common(10)
```

```python
common_words = pd.DataFrame(freq.most_common(10), columns=["word", "count"])

plt.figure(figsize=(8, 4))
sns.barplot(data=common_words, x="count", y="word", color="#2563eb")
plt.title("Most Common Words in Student Responses")
plt.show()
```

## 8. Sentiment Analysis

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">What is sentiment analysis?</h2>
Sentiment analysis tries to measure whether a text is positive, negative, or neutral. It is useful in social science for analyzing public opinion, policy responses, reviews, and social media discussions.<br><br>
A sentiment score is not the same as full interpretation, but it can provide a useful summary signal.
</div>

```python
sia = SentimentIntensityAnalyzer()

sentiment_scores = []
for response in responses:
    scores = sia.polarity_scores(response)
    sentiment_scores.append({"text": response, **scores})

sentiment_df = pd.DataFrame(sentiment_scores)
sentiment_df[["text", "neg", "neu", "pos", "compound"]]
```

## 10. Deep Learning

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">What is deep learning?</h2>
Deep learning is a family of machine learning methods based on neural networks with many layers. These models can learn complex patterns from large amounts of data.<br><br>
In NLP, deep learning became especially important because language has many layers of structure: word order, context, meaning, syntax, and ambiguity. Traditional methods can capture some of this, but deep learning often captures much more.
</div>

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">When is deep learning used?</h2>
Deep learning is especially useful when:
<ul>
<li>the data are large and complex</li>
<li>patterns are difficult to hand-code</li>
<li>you are working with images, audio, or text</li>
<li>you care about predictive performance and representation learning</li>
</ul>
Examples include image recognition, speech recognition, language modeling, translation, and text generation.
</div>

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">Pros and cons of deep learning</h2>
<b>Pros</b>
<ul>
<li>very powerful for complex pattern recognition</li>
<li>works well with images, text, and audio</li>
<li>can automatically learn useful features from raw data</li>
</ul>
<b>Cons</b>
<ul>
<li>often needs a lot of data</li>
<li>can require substantial computing power</li>
<li>harder to interpret than simpler models</li>
<li>can reproduce bias in training data</li>
</ul>
</div>

## 11. Neural Network Intuition

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">How a neural network works</h2>
A neural network takes inputs, transforms them through one or more hidden layers, and produces an output. Each layer learns a representation of the data. In a simple view:
<ol>
<li>Inputs enter the network</li>
<li>Weighted sums are computed</li>
<li>An activation function introduces nonlinearity</li>
<li>The network compares predictions to true values</li>
<li>The model updates its weights through backpropagation</li>
</ol>
</div>

## 12. Implementing Deep Learning in Python

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">Common Python libraries</h2>
Deep learning in Python is usually done with libraries such as:
<ul>
<li>TensorFlow / Keras</li>
<li>PyTorch</li>
</ul>
Below is a simple example using Keras. The goal here is not to master every detail, but to see what a deep learning workflow looks like in code.
</div>

```python
# Example code: a very small neural network for binary classification
# This cell may require TensorFlow to be installed in your environment.

example_code = '''
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

X = np.array([[0.1, 0.3], [0.2, 0.4], [0.8, 0.9], [0.7, 0.6]])
y = np.array([0, 0, 1, 1])

model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=20, verbose=0)
predictions = model.predict(X)
print(predictions)
'''

print(example_code)
```

## 13. Introduction to LLMs and AI

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">What is a large language model?</h2>
A large language model (LLM) is a deep learning model trained on massive amounts of text. Its task is usually to predict the next token in a sequence, but in doing so it learns many useful language patterns. This is why LLMs can answer questions, summarize text, generate writing, translate, and hold conversations.<br><br>
LLMs are part of AI, but they are not the whole of AI. They are one powerful family of models within a larger field.
</div>

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">How LLMs connect to earlier concepts</h2>
The progression in this notebook is important:
<ul>
<li>NLTK shows classic NLP techniques such as tokenization and frequency analysis</li>
<li>Deep learning introduces neural networks and representation learning</li>
<li>LLMs combine deep learning with very large-scale text training to model language in flexible ways</li>
</ul>
In other words, LLMs build on both NLP and deep learning.
</div>

### Real-world uses of LLMs

- summarizing research articles
- drafting code or writing
- answering questions from documents
- tutoring and educational support
- analyzing large text collections
- chatbots and digital assistants

### Important risks

- hallucinations or fabricated information
- bias in outputs
- privacy concerns
- overreliance by users
- unclear authorship or accountability
