
## 3. Lists

A **list** is a collection of items in a single variable.

- Items can be **different types**: numbers, strings, booleans, even other lists.  
- Lists are **ordered**, meaning the items keep their position, and you can access them by **index** (starting from 0).  
- Lists are **mutable**, meaning you can **change, add, or remove items** after creating them.

## Examples

```python
# 1️⃣ Simple list of fruits
fruits = ["apple", "banana", "cherry"]
print(fruits)

# 2️⃣ List with mixed types
my_list = [10, "hello", True, 3.14]
print(my_list)

# 3️⃣ Access items by index
print(fruits[0])  # first item -> apple
print(fruits[2])  # third item -> cherry

# 4️⃣ Slice a list
print(fruits[0:2])  # first two items -> ['apple', 'banana']

# 5️⃣ Modify a list
fruits[1] = "blueberry"  # replace 'banana'
fruits.append("orange")   # add a new item
fruits.insert(1, "kiwi") # insert 'kiwi' at index 1
print(fruits)

# 6️⃣ Remove items
fruits.pop(0)            # remove first item
fruits.remove("orange")  # remove 'orange'
print(fruits)

## 4. Dictionaries

A **dictionary** is a collection of **key-value pairs** stored in a single variable.

- Each **key** maps to a **value**.
- Dictionaries are **unordered** (Python 3.7+ maintains insertion order) and **mutable**.
- Keys must be **unique** and **immutable** (strings, numbers, tuples, etc.).
- Values can be **any data type**, including numbers, strings, lists, or even other dictionaries.

---

## Examples

```python
# 1️⃣ Creating a dictionary
student_scores = {"Alice": 90, "Bob": 85, "Charlie": 92}
print(student_scores)

# 2️⃣ Accessing values by key
print(student_scores["Alice"])  # Output: 90

# 3️⃣ Adding or updating a key-value pair
student_scores["David"] = 88       # add new
student_scores["Bob"] = 95         # update existing
print(student_scores)

# 4️⃣ Removing a key-value pair
del student_scores["Charlie"]
print(student_scores)

# 5️⃣ Nested dictionary
nested_dict = {"Class1": {"Alice": 90, "Bob": 85}, "Class2": {"Charlie": 92}}
print(nested_dict["Class1"]["Bob"])  # Output: 85

## 6. Tuples

A **tuple** is a collection of items, just like a list, but **immutable**, meaning you **cannot change, add, or remove items** after creating it.  

- Items can be **different types**: numbers, strings, booleans, etc.  
- Tuples are **ordered**, so you can access items by **index**.  
- Tuples are defined using **parentheses `()`**.
- This is useful when you want to **protect data** from accidental modification.

### Examples:

```python
# 1️⃣ Simple tuple
my_tuple = (1, 2, 3, 4)
print(my_tuple)

# 2️⃣ Mixed types
mixed = (10, "hello", True)
print(mixed)

# 3️⃣ Access items by index
print(my_tuple[0])  # 1
print(mixed[2])     # True


## 7. Sets

A **set** is an **unordered collection of unique items**.  

- Items are **unique** → no duplicates allowed.  
- Sets are **unordered** → you cannot access items by index.  
- Sets are **mutable** → you can add or remove items.  
- Defined using **curly braces `{}`** or the `set()` function.  

### Examples:

```python
# 1️⃣ Create a set
my_set = {1, 2, 3, 4}
print(my_set)

# 2️⃣ Set with duplicate items
duplicates = {1, 2, 2, 3}
print(duplicates)  # Output: {1, 2, 3}

# 3️⃣ Add and remove items
my_set.add(5)        # add a new item
my_set.remove(2)     # remove an item
print(my_set)

# 4️⃣ Operations between sets
a = {1, 2, 3}
b = {3, 4, 5}

print(a.difference(b))  # {1, 2}

# Comparison Operators in Python

**Comparison operators** are used to compare values.  
They return a **Boolean value**: `True` or `False`.  

| Operator | Meaning                  | Example          | Result  |
|----------|-------------------------|----------------|---------|
| `==`     | Equal to                | `5 == 5`       | True    |
| `!=`     | Not equal to            | `5 != 3`       | True    |
| `>`      | Greater than            | `5 > 3`        | True    |
| `<`      | Less than               | `5 < 3`        | False   |
| `>=`     | Greater than or equal   | `5 >= 5`       | True    |
| `<=`     | Less than or equal      | `5 <= 3`       | False   |

### Examples:

```python
x = 10
y = 7

print(x == y)   # False
print(x != y)   # True
print(x > y)    # True
print(x < y)    # False
print(x >= 10)  # True
print(y <= 5)   # False

# Logical Operators in Python

**Logical operators** combine **Boolean values** (`True` or `False`) to produce a new Boolean result.  

| Operator | Meaning                       | Example                     | Result  |
|----------|-------------------------------|-----------------------------|---------|
| `and`    | True if **both** values True  | `True and False`            | False   |
| `or`     | True if **at least one** True | `True or False`             | True    |


### Examples:

```python

# and
(1 > 2) and (2 < 3)   # False

# or
(1 > 2) or (2 < 3)    # True

(1 == 2) or (2 == 3) or (4 == 4)    # True

# Conditional Statements (`if`, `elif`, `else`)

Conditional statements let your program **make decisions** based on **Boolean conditions**.  

- **`if`** → runs a block of code if the condition is `True`  
- **`elif`** → (else if) runs a block if the previous `if` or `elif` was `False` and this condition is `True`  
- **`else`** → runs a block if **none of the previous conditions** were `True`  

### Example 1: Simple if-else

```python
x = 10

if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")

# For Loops

A **for loop** is used to **repeat a block of code** for each item in a sequence (like a list, tuple, or string).  

- It is useful when you want to **perform the same operation multiple times**.  
- The loop **iterates** over each item in the sequence automatically.  

### Example: Printing student names

```python

    print("Hello,", student)

# While Loops

A **while loop** repeats a block of code **as long as a condition is True**.  

- Use `while` when you **don’t know in advance how many times the loop will run**.  
- Make sure the condition will eventually become `False`, or the loop will run forever!  

### Example: Counting down student attempts

```python
attempts = 3

while attempts > 0:
    print("You have", attempts, "attempt(s) left to submit your assignment.")
    attempts -= 1

print("No attempts left. Please ask your instructor for help.")

<div style="background-color: #eaf7ea; padding: 15px; border-radius: 10px; border-left: 5px solid #4caf50;">

# List Comprehension

A **list comprehension** is a concise way to **create a new list** by applying an expression to each item in an existing list.  
It’s equivalent to a `for` loop, but shorter and often easier to read.

### Example: Square each number in a list

```python
x = [1, 2, 3, 4]

# Using a for loop
out = []
for item in x:
    out.append(item**2)
print(out)

# Using a list comprehension
out2 = [item**2 for item in x]
print(out2)

## 📝 Try It Yourself 

Define a function called `grade_students` that:

1. Takes a **list of student scores** as a parameter.  
2. Prints each score with a message:  
   `"Student <number> scored <score>"`.  
3. If a score is below 70, prints `"Needs extra help"`.  
4. Call the function with a **list of at least 5 scores**.

````python

# Define the function
def grade_students(scores):
    """
    Takes a list of student scores.
    Prints each score with the student number.
    Flags scores below 70.
    """
    for i, score in enumerate(scores, start=1):
        print(f"Student {i} scored {score}")
        if score < 70:
            print("Needs extra help")

# Example list of scores
student_scores = [88, 92, 75, 60, 95]

# Call the function
grade_students(student_scores)
````

# Introduction to Pandas

**Pandas** is a library used for **data analysis and data manipulation**.

Its main structure is called a **DataFrame**, which looks like a spreadsheet or table.

DataFrames allow us to:

- load datasets
- clean messy data
- filter rows
- compute statistics


````python

import pandas as pd

data = {
    "name": ["Alice","Bob","Charlie"],
    "age": [24, 30, 22]
}

df = pd.DataFrame(data)

print(df)
print(df.describe())
````
