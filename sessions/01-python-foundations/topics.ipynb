## 1. Data Types
Python has several built-in types. We'll focus on the ones you use constantly in data science.


### Numbers
Integers (`int`) are whole numbers and floats (`float`) have decimals.


```python
x = 10
pi = 3.14159
print(type(x), x)
print(type(pi), pi)
```

### Strings
Strings (`str`) store text. Use quotes and string operations for formatting.


```python
name = "Ada"
message = f"Hello, {name}!"
print(message)
print(message.upper())
```

### Printing
`print()` shows output. Great for debugging and explanations.


```python
print("This is a printed message.")
```

### Lists
Lists store ordered collections and are mutable.


```python
cities = ["Nairobi", "Accra", "Cairo"]
print(cities)
print(cities[0])
```

### Dictionaries
Dictionaries store key-value pairs.


```python
student = {"name": "Sam", "year": 2, "major": "Biology"}
print(student["name"])  # lookup by key
```

### Booleans
Booleans (`bool`) are `True` or `False`.


```python
is_student = True
print(is_student, type(is_student))
```

### Tuples
Tuples are ordered and immutable.


```python
point = (3, 4)
print(point)
```

### Sets
Sets store unique values and are unordered.


```python
colors = {"red", "blue", "red"}
print(colors)
```

## 2. Comparison Operators
Used to compare values and return booleans.


```python
print(5 > 3)
print(5 == 3)
print(5 != 3)
```

## 3. `if`, `elif`, `else`
Control the flow of your code.


```python
temp = 72
if temp > 80:
    print("It's hot")
elif temp > 60:
    print("It's nice")
else:
    print("It's chilly")
```

## 4. `for` Loops
Iterate over items in a list.


```python
for city in cities:
    print(city)
```

## 5. `while` Loops
Repeat until a condition is false.


```python
count = 0
while count < 3:
    print("count:", count)
    count += 1
```

## 6. `range()`
Generates a sequence of numbers.


```python
for i in range(5):
    print(i)
```

## 7. List Comprehension
Compact way to build lists.


```python
squares = [n**2 for n in range(1, 6)]
print(squares)
```

## 8. Functions
Functions make your code reusable.


```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Rosa"))
```

## 9. Lambda Expressions
Small anonymous functions.


```python
double = lambda x: x * 2
print(double(4))
```

## 10. `map` and `filter`
Apply a function to a sequence or filter items.


```python
nums = [1, 2, 3, 4, 5]

mapped = list(map(lambda x: x * 10, nums))
filtered = list(filter(lambda x: x % 2 == 0, nums))

print(mapped)
print(filtered)
```

## 11. Methods
Methods are functions attached to objects.


```python
text = "data science"
print(text.title())
print(text.replace("science", "fun"))
```
