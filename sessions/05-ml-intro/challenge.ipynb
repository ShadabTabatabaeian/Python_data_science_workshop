## Context

Artificial intelligence tools such as ChatGPT, Copilot, Gemini, Claude, Midjourney, and Grok are changing how people work. Organizations are using AI to automate tasks, improve productivity, and support decision-making.

This dataset contains 30,000 synthetic but realistic employee records across multiple industries and job roles. It is clean and has no missing values.

Kaggle source: [AI Tools Usage and Workplace Productivity Dataset](https://www.kaggle.com/datasets/shree0910/ai-tools-usage-and-workplace-productivity-dataset)

In this project, you will use the dataset to answer a focused question:

> Can we predict whether an employee received a promotion in the last year based on their work conditions, AI use, and career profile?

This is a good question for **logistic regression** because the Outcome measure, `Promotion_Last_Year`, has two categories: `Yes` and `No`.

## Dataset Features

Some important variables in the dataset include:

- `Age`
- `Industry`
- `Job_Role`
- `Experience_Years`
- `Company_Size`
- `Remote_Work_Type`
- `AI_Tools_Used`
- `Daily_AI_Usage_Hours`
- `Primary_AI_Tool`
- `Tasks_Automated_Percent`
- `Time_Saved_Per_Day_Minutes`
- `Learning_Time_Per_Week_Hours`
- `Meetings_Per_Day`
- `Work_Hours_Per_Day`
- `Job_Satisfaction`
- `Productivity_Change`
- `Salary_USD`
- `Promotion_Last_Year`
- `Fear_of_AI_Replacing_Job`

There are multiple possible Outcome measures in this dataset. For this project, we will focus on `Promotion_Last_Year`.

<div style="background-color:#e6f2ff; padding:16px; border-radius:12px; border-left:6px solid #3b82f6;">
<h2 style="margin-top:0;">Why logistic regression?</h2>
Logistic regression is used when the outcome variable is binary. In this dataset, <code>Promotion_Last_Year</code> is either <code>Yes</code> or <code>No</code>. That makes logistic regression a natural first model.<br><br>
Substantively, this means the model can help us ask a social science question about inequality and work: which features of employees and workplaces are associated with higher odds of promotion?
</div>
