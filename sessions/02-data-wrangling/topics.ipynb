## Introduction to NumPy

NumPy (Numerical Python) is a library used for working with numbers efficiently.

It is especially useful for:

- Performing calculations on large datasets  
- Computing statistics  
- Handling numerical arrays  



Numpy allows you to work with **arrays with objects all of the same type** (all int, or all bool, etc.).

So you can do computations very efficiently without having to use loops. 

</div>

## Creating a numpy.ndarray

The basic data type of numpy is an **ndarray**.

ndarray = **N-dimensional array.**

A simple way to create an ndarray is np.arange (“a range”).


````python
import numpy as np

# Works similar to range(N)
np.arange(1, 4)

## step size = 2
np.arange(1, 4, step = 2)
````

### Descriptive Statistics with NumPy

We often want to summarize data using:
- Mean (average)
- Standard deviation (spread)
- Sum (total)

````python
# Imagine these are responses to a Likert-scale question (1–5)
responses = np.array([1, 2, 3, 4, 5, 3, 4, 2, 1, 5])
responses

responses.sum()   # Total sum
responses.mean()  # Average
responses.std()   # Standard deviation
responses.min()   # Minimum
responses.max()   # Maximum
````

# Introduction to Pandas

**Pandas** is used for working with tabular data (like Excel or CSV files).

It allows you to:
- Organize data into rows and columns
- Filter and select data
- Compute statistics easily

The main structure is called a **DataFrame**.

</div>

````python

import pandas as pd
````

### Creating a DataFrame (Example: student dataset)

````python
data = {
    "age": [18, 19, 20, 21, 22],
    "study_hours": [2, 3, 5, 4, 6],
    "score": [70, 75, 85, 80, 90]
}

df = pd.DataFrame(data)
df
````

### Exploring the Data

We can quickly summarize the dataset:


````python
df.describe()
````

### Selecting Data

We often want to focus on a specific variable:


````python
df['study_hours']
````

### Filtering Data

Example: students who studied more than 3 hours

````python
df[df['study_hours'] > 3]
````

### Why this matters

In social science research, you might:
- Filter participants based on conditions
- Compare subgroups
- Explore relationships between variables

## Data Wrangling (Cleaning Data)

Real-world data is messy. You will often encounter:
- Missing values
- Incomplete responses

Pandas provides tools to handle this.


````python
df2 = pd.DataFrame({
    "age": [18, None, 20, 21],
    "score": [70, 75, None, 80]
})

df2
````

### Sorting
Use `sort_values`.


````python
df.sort_values("age", ascending=False)
````

### Groupby and Aggregation
Summarize by groups. Why isn't the example below a good option for grouping?

````python
df.groupby("score")["age"].mean()
````

### Dropping Missing Values

````python
df2.dropna()
````

### Filling Missing Values

````python
df2.fillna(df2.mean(numeric_only=True))
````

### Why this matters

Decisions about missing data can affect your results:
- Dropping data reduces sample size
- Filling data introduces assumptions

These are important research decisions.


````python
# more useful methods

# Show all unique values present in the 'age' column
df2['age'].unique()

# Count how many distinct values exist in the 'age' column
df2['age'].nunique()

# Count how many times each unique value appears in the 'age' column
df2['age'].value_counts()
````

````python
#Select from DataFrame using criteria from multiple columns
newdf = df2[(df2['age']>20) & (df2['age']==21)]

print(newdf)
````
