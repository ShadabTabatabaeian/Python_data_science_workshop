## 1. Meet pandas
pandas provides two core data structures: **Series** (1D) and **DataFrame** (2D).


```python
import pandas as pd
```

## 2. Creating a DataFrame
We'll create a small dataset by hand.


```python
data = {
    "student": ["Amina", "Jordan", "Kai", "Priya", "Luis"],
    "major": ["Bio", "CS", "Math", "Econ", "Bio"],
    "gpa": [3.6, 3.9, 3.2, 3.7, None],
    "credits": [60, 75, 45, 90, 30]
}

df = pd.DataFrame(data)
df
```

## 3. Selecting Columns
Use bracket notation for columns.


```python
df["gpa"]
```

## 4. Filtering Rows
Use boolean conditions.


```python
df[df["credits"] >= 60]
```

## 5. Sorting
Use `sort_values`.


```python
df.sort_values("gpa", ascending=False)
```

## 6. Groupby and Aggregation
Summarize by groups.


```python
df.groupby("major")["gpa"].mean()
```

## 7. Missing Values
Detect and fill missing values.


```python
df["gpa"].isna()
```

```python
df["gpa"].fillna(df["gpa"].mean())
```
